```python
from litellm import completion
from pathlib import Path
import sys
import requests

# Make the project importable from the notebook
root_dir = Path.cwd().resolve().parents[0]
sys.path.insert(0, str(root_dir / "src"))

from deal_hunter.services.preprocessing import Preprocessor

# 1) LLM quick sanity check
messages = [{"role": "user", "content": "Hello how are you"}]
response = completion(
    messages=messages,
    model="groq/openai/gpt-oss-20b",
    reasoning_effort="low",
)
print(response.choices[0].message.content)


# 2) IP lookup
info = requests.get("https://ipinfo.io/json").json()
city, region, country, ip = info["city"], info["region"], info["country"], info["ip"]
print(f"{city} {region} {country} {ip}")
```

```python
# 2) Preprocess sample
raw_text = (
    "Apple 2025 MacBook Air (13-inch, Apple M4 chip with 10-core CPU and 8-core GPU, "
    "16GB Unified Memory, 256GB) - Starlight"
)
preprocessor = Preprocessor()
processed = preprocessor.preprocess(raw_text)
print(processed)
```
